## Задача

Даны две функции: 
f(x) = x^3 - 50x и g(x) = -x^4 + 88x^2 - 241

### Требуется:

Найти координаты точек пересечения


Построить графики функций в одной системе координат


Построить график функции пересечения

## Решение

Для нахождения точек пересечения приравниваем две функции друг к другу

получаем функцию пересечения fgx (см. ниже)

```python
from sympy import *
x = Symbol('x')
fx = x**3 - 50*x
gx = -x**4 + 88*x**2 - 241
fgx = fx-gx
fgx
```

### Корни

```python
roots = real_roots(fgx)
roots[0].evalf()
```

```python
cross_dict = {}
c = 1
for z in roots:
    cross_dict[c]={'x': round(N(z), 5), 'y': round(N(fgx.evalf(subs={x: z}, chop=True)), 5)}
    c += 1

cross_dict
```

```python
plot(fgx,x)
```
